##[Independence](https://en.wikipedia.org/wiki/Independence_%28probability_theory%29)

Only independent if joint probabilities = product of their probabilities

$$Pr(x,y) = Pr(x)Pr(y)$$

---
$$Pr(X|Y) = Pr(X)$$
$$Pr(Y|X) = Pr(Y)$$

- Based on linear independence and conditional probabilites

n events
- A finite set of events is mutually independent if every event is independent of any intersection of the other events
- multiplication rule for independent events

$$Pr(x,y,z) = Pr(x)Pr(y)Pr(z)$$
or
$$P(B_i\cap B_{i+1} \cap... B_k) = \prod_{i=1}^{k}P(B_i)$$

#Expectation
- short average or mean
Example

```
import numpy as np

X = np.array([1,2,3,4,5,6])
P = np.array([1/6,1/6,1/6,1/6,1/6,1/6])
sum =0
for i in range(len(X)):
  sum+=(X[i]*P[i])
print(sum)

```

##Expectation Properties
$$1. E[k]=k$$
$$2. E[kx]=kE[x]$$
$$3. E[E[x]]=E[x]$$
$$4. E[x+y]=E[x]+E[y]$$
$$5. E[xy]=E[x]E[y]$$

- Useful when dealing with other statistical measurements such as variance

##[Variance](https://en.wikipedia.org/wiki/Variance)
$$Var(X) = E[(X-ux)^2]$$
 $$=E[X^2]-E[x]^2$$

 Using properties shows that variance could be expressed as subtracting squared version of a normal expectation from expected value of random variable squared

 ##[Covariance](https://en.wikipedia.org/wiki/Covariance)
 - Joint variablity of two random variables
 - adding second RV
 - If X and Y are indepenant of each other
 $$Cov(X,Y) = E[(X-ux)(Y-ux)]$$
 $$= E[XY] - E[X]E[Y]$$
 - therefore 
 $$Cov(X,Y) = 0$$

 - Basically the idea that if there is some sort of trend in the data that can relate the two variables to one another their Covariance is not 0

##[Skewness](https://en.wikipedia.org/wiki/Skewness)
$$Skew[X] = E[(\frac{X-ux}{\sigma})^3]$$
- Tells how the distribution is leaning  towards one side (right or left)

##[Kurtosis](https://en.wikipedia.org/wiki/Kurtosis)
$$Kurt[X] = E[(\frac{X-ux}{\sigma})^4]$$
- Measure of sharpness of the distribution 
- Higher the kurtosis the more sharp the peak is
