##Marginalization
-The way to ignore a certain random variables to get a probability with less random variables.
- Calculate single probability from joint probabilities
- Works for n variables that have corilation 
1. Sum up probabilities for known variables for variable condition you want to find

$$Pr(x) = \int Pr(x,y) \,dy  ...(continious)$$

$$Pr(x) = \sum Pr(x,y) ... (Discrete)$$

$$Pr(x=0) = \sum_{y=0}^{1} Pr(x=0,y)$$

$$ = Pr(x=0,y=0) + Pr(x=0,y=1)$$ 

2. If variable z was introduced into this problem where x - y - z we would have

$$Pr(x=0) = \sum_{z=0}^{1}\sum_{y=0}^{1}Pr(x=0,y,z)$$

3. key for marginalization is to understand the graph and the variable relationships




```
##Prob X
X1 = 0.6
##X0 unknown
Y1 = 0.3
Y0 = 0.7

##Joint probabilities
X0Y0 = 0.28 ##given
X0Y1 = 0.12 ##given
X1Y0 = X1*Y0
X1Y1 = X1*Y1

##How to get Pr(x=0)
X0 = X0Y0+X0Y1
##should get 0.4
print(X0)
```

##[Conditional Probability](https://en.wikipedia.org/wiki/Conditional_probability)

- Consider each state as independant
- The probability of an event given that (by assumption, presumption, assertion or evidence) another event has occurred.
1. Given y, Prob of X
2. Using joint probability of X and Y over the normalization of Y

$$Pr(X|Y) = \frac{Pr(X,Y)}{Pr(Y)}$$



##[Bayes Rule](https://en.wikipedia.org/wiki/Bayes%27_theorem)

-The rule that ties posterior probability with likelihood, prior, and evidence.

$$Pr(y|x) = \frac{Pr(x|y) Pr(y)}{Pr(x)}$$

1. Works for n variables that are indepenant of eachother based on the conditional probability rule
